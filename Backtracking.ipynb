# Permutations

This function uses backtracking to generate permutations of a set.

```python
def permutations(s, perm=None):
  if perm is None:
    perm = list()

  if len(perm) == len(s):
    yield perm.copy() #yeild copy to avoid aliasing
  else:
    for value in s - set(perm):
      perm.append(value)
      for complete_perm in permutations(s, perm):
        yield complete_perm
      perm.pop() # Back up
```

```python
for perm in permutations({1, 2, 3,4,5,6,7,8,9}):
  print(perm)
```

# Combinations

This function uses backtracking to generate combinations of a set.

```python
def combinations(s, k, combo=None, order=None, index=None):
  if combo is None:
    combo = set()
    order = list(s) # Ordered Version of entire set
    index = 0

  if len(combo) == k:
    yield combo.copy() #yeild copy to avoid aliasing
  else:
    for i in range(index, len(order)):
      value = order[i]
      combo.add(value)
      for complete_combo in combinations(s, k, combo, order, i+1):
        yield complete_combo
      combo.remove(value) # Back up
```

```python
for combo in combinations({1, 4, 3, 2}, 2):
  print(combo)
```
