# COMPAS

```r
# Download the COMPAS dataset from ProPublica
download.file("https://raw.githubusercontent.com/propublica/compas-analysis/master/compas-scores-two-years.csv", "Compas.csv")
Compas <- read.csv("Compas.csv", stringsAsFactors = T)
head(Compas)
```

```r
# Drop rows with too long between arrest and screening
Compas <- Compas[Compas$days_b_screening_arrest <= 30,]
Compas <- Compas[Compas$days_b_screening_arrest >= -30,]
```

```r
# Add some binary variables for risk categories
Compas$low_risk <- Compas$score_text == "Low"
Compas$high_risk <- Compas$score_text == "High"
```

```r
# Pull out some subgroups
White <- Compas[Compas$race == "Caucasian",]
Black <- Compas[Compas$race == "African-American",]
```

# Tables

```r
# Race vs low-risk for all inmates
table(Compas$race, Compas$low_risk)
```

```r
# Recidivism vs low-risk for white inmates
table(White$two_year_recid, White$low_risk)
```

```r
# Recidivism vs low-risk for black inmates
table(Black$two_year_recid, Black$low_risk)
```
