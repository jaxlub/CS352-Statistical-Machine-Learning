```python
from transformers import pipeline
```

```python
gpt1 = pipeline("text-generation", model="openai-community/openai-gpt")
```

```python
gpt2 = pipeline("text-generation", model="openai-community/gpt2-large")
```

```python
gpt1("I love when you")
```

```python
gpt2("I love when you", max_new_tokens=50, top_k=10, penalty_alpha = .6)
```
