# Linear regression

```r
# Download the Advertising data
download.file("https://myslu.stlawu.edu/~ltorrey/ml/Advertising.csv", "Advertising.csv")
Advertising <- read.csv("Advertising.csv")
```

```r
# build linear model
TV.lm <- lm(sales ~ TV, data = Advertising)
```

```r
# see how it looks
plot(Advertising$TV, Advertising$sales)
abline(TV.lm, col = "blue")
```

```r
summary(TV.lm)
```

# Multiple linear regression

```r
# build a full linear model
full.lm <- lm(sales ~ TV + radio + newspaper, data = Advertising)
summary(full.lm)
```

```r
reduced.lm <- lm(sales ~ TV + radio, data = Advertising)
summary(reduced.lm)
```

```r
# Adding an interaction term
interaction.lm <- lm(sales ~ TV + radio + TV:radio, data = Advertising)
summary(interaction.lm)
```
