# Cancer

```python
import pandas as pd
```

```python
# Download the dataset
cancer = pd.read_csv("https://myslu.stlawu.edu/~ltorrey/ml/Cancer.csv")
```

```python
# Separate features and target
x = cancer.drop("diagnosis", axis=1)
y = cancer["diagnosis"]
```

# Linear SVM

```python
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
# Baseline accuracy
svm = make_pipeline(StandardScaler(),LinearSVC())
cross_val_score(svm,x,y).mean()
```

```python
# Settings to Try
hyperparameters = {
    "linearsvc__C": [0.01,0.1,1,10,100],
    "linearsvc__max_iter":[100000]
}
# Tuned accuracy
tuned_svm = GridSearchCV(svm, hyperparameters)
cross_val_score(tuned_svm,x,y).mean()
```

# Synthetic data

```python
from sklearn.datasets import make_circles
import seaborn as sns
```

```python
# Make non-linear dataset
x,y = make_circles()
plot = sns.scatterplot(x=x[:,0],y=x[:,1], hue = y)
```

# Kernel SVMs

```python
from sklearn.svm import SVC
```

```python
# Linear Kernal
linear_svm = SVC(kernel = "linear")
cross_val_score(linear_svm,x,y).mean()

```

```python
# Polynomial Kernel
poly_svm = SVC(kernel="poly", degree = 2)
cross_val_score(poly_svm,x,y).mean()
```

```python
# RBF Kernel (similar to similarity features)
rbf_svc = SVC(kernel = "rbf")
cross_val_score(rbf_svc,x,y).mean()
```
